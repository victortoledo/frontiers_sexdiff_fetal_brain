#### Statistical analysis of covariates
##### These tests are relevant to demonstrate that the covariates in the two analysed groups are similar, and thereby the observed differences are not due to these covariates
#  
##### Is sex related to the batch (read length)?

```r
## Preparing the covariates table:
covariates$ReadLength <- as.factor(covariates$ReadLength)
covariates$Sex <- as.factor(covariates$Sex)

### Categorical variables:
#print("Is sex related to the batch (read length)?")
chisq.test(covariates$Sex,covariates$ReadLength)
```

##### Is the distribution of the RIN value normal in female samples?

```r
### Continuous variables:

#RIN
#print("Is the distribution of the RIN value normal in female samples?")
shapiro.test(covariates$RIN[covariates$Sex == "Female"])
```

##### Is the distribution of the RIN value normal in male samples?

```r
#print("Is the distribution of the RIN value normal in male samples?")
shapiro.test(covariates$RIN[covariates$Sex == "Male"])
```

##### Is sex related to the RIN values?

```r
#print("Is sex related to the RIN values?")
t.test(covariates$RIN ~ covariates$Sex)
```

##### What is the standand deviation of the RIN values in female samples?

```r
#print("What is the standand deviation of the RIN values in female samples?")
sd(covariates$RIN[covariates$Sex == "Female"])
```

##### What is the standand deviation of the RIN values in male samples?

```r
#print("What is the standand deviation of the RIN values in male samples?")
sd(covariates$RIN[covariates$Sex == "Male"])
```

##### Is the distribution of age normal in female samples?

```r
#PCW
#print("Is the distribution of age normal in female samples?")
shapiro.test(covariates$PCW[covariates$Sex == "Female"])
```

##### Is the distribution of age normal in male samples?

```r
#print("Is the distribution of age normal in male samples?")
shapiro.test(covariates$PCW[covariates$Sex == "Male"])
```

##### Is sex related to the age?

```r
#print("Is sex related to the age?")
t.test(covariates$PCW ~ covariates$Sex)
```

##### From these results, we can delete 16428 because it's an outlier
###### Anyway, let's follow with a PCA to see any other possible outliers, starting with a screeplot to see how much of the data each principal component explains

```r
#Running a PCA to see if the same sample appears as an outlier:
res_pca_tpm = prcomp(t(counts), scale = TRUE, center = TRUE)
summary_pca_tpm = summary(res_pca_tpm)
eigval_tpm = get_eigenvalue(res_pca_tpm)

# PCA:
# First the screeplot containing the percentage of explained variance by each principal component:
eigval_tpm$dimensions = 1:nrow(eigval_tpm)
eigval_tpm = eigval_tpm[1:10,]


scree <- ggplot(eigval_tpm, aes(x = dimensions, y = variance.percent, fill = "#6495ED")) +
  geom_bar(stat = "identity") +
  labs(title= "Screeplot - Eigenvalues of PCA", x="Dimensions", y = "Percentage of explained variance (%)") +
  guides(fill = FALSE) +
  scale_y_continuous(breaks=seq(0, 35, 5)) +
  theme_grey() +
  scale_x_continuous(breaks=seq(0, 10, 1)) +
  geom_line(color="gray30") +
  geom_point(color="gray30") +
  theme(plot.title = element_text(size = 14, hjust = 0.5),
        text = element_text(size = 12))
scree

prefix <- file.path('~/Documents/projects/ms-victor/sex_differences_fetal_brain/preprocessing/_m/screeplot_ggplot.')
  devices <- c('pdf', 'jpeg', 'png')
  walk(devices,
     ~ ggsave(filename = file.path(paste0(prefix, .x)), device = .x))
```

##### The next step will be running a singular value decomposition (SVD):

```r
#this section loads a function and should be unchanged
#loading a colourblind pallete

cbPalette <- c("#999999", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")

mySVDplot <- function (PhenoTypes.lv, plan, resultsDir, filename){
  ### Phenotypes.lv : sample information
  ### plan : normalized counts
  ### resultDir : results folder
  ### Calculate SVD
  #https://feliperego.github.io/blog/2015/10/23/Interpreting-Model-Output-In-R
  svd(as.matrix(plan))-> svd.o
  
  topPCA <- 10
  svdPV.m <- matrix(nrow=topPCA,ncol=ncol(PhenoTypes.lv));
  colnames(svdPV.m) <- colnames(PhenoTypes.lv);

  for(c in 1:topPCA){
    for(f in 1:ncol(PhenoTypes.lv)){
      if(class(PhenoTypes.lv[,f])!="numeric") {
        svdPV.m[c,f] <- kruskal.test(svd.o$v[,c] ~ as.factor(PhenoTypes.lv[[f]]))$p.value
        } else {
          svdPV.m[c,f] <- summary(lm(svd.o$v[,c] ~ PhenoTypes.lv[[f]]))$coeff[2,4]
        }
    }
  }

  write.table(svdPV.m, paste(resultsDir , "/", filename, ".txt", sep=""), sep="\t", row.names = F, quote=F)

  ######################## Plot SVD Image #############################
  myPalette <- c("#4B3D60","#FD5E53","#FC9C54","#FFE373","white");
  breaks.v <- c(-200,-10,-5,-2,log10(0.05),0);

  par(mar=c(5,15,2,1));
  image(x=1:nrow(svdPV.m), 
        y=1:ncol(svdPV.m), 
        z=log10(svdPV.m), 
        col=myPalette, 
        breaks=breaks.v, 
        xlab="", 
        ylab="", 
        axes=FALSE, 
        main= "SVD analysis");

  axis(1,at=1:nrow(svdPV.m),
       labels=paste("PC-",1:nrow(svdPV.m),sep=""),
       las=2);
  suppressWarnings(axis(2,at=1:ncol(svdPV.m),
                        labels=str_wrap(as.character(colnames(svdPV.m)), 
                                        width = 40),las=2));
  legend(x=-(topPCA/2.5),
         y=3,
         legend=c(expression("p < 1x"~10^{-10}),
         expression("p < 1x"~10^{-5}),"p < 0.01", "p < 0.05", "p > 0.05"),
         fill=c("#4B3D60","#FD5E53","#FC9C54","#FFE373","white"),
         par('usr')[2], par('usr')[4], xpd=NA);
  
  ### PDF saving
  pdf(paste(resultsDir,"/", filename, ".pdf",sep=""),width=8,height=8);
  par(mar=c(5,15,2,1),xpd=TRUE);
  image(x=1:nrow(svdPV.m), 
        y=1:ncol(svdPV.m), 
        z=log10(svdPV.m), 
        col=myPalette, 
        breaks=breaks.v, 
        xlab="", 
        ylab="", 
        axes=FALSE, 
        main= "SVD analysis");
  axis(1,at=1:nrow(svdPV.m),
       labels=paste("PC-",1:nrow(svdPV.m),sep=""),
       las=2);
  suppressWarnings(axis(2,at=1:ncol(svdPV.m),
                  labels=str_wrap(colnames(svdPV.m), width = 30),
                  las=2));
  #abline(h=c(1:ncol(svdPV.m)+0.5), col="cornsilk3", lty=3)
  legend(x=-(topPCA/2.5),
         y=3,legend=c(expression("p < 1x"~10^{-10}),
        expression("p < 1x"~10^{-5}),"p < 0.01", "p < 0.05", "p > 0.05"), 
        fill=c("#4B3D60","#FD5E53","#FC9C54","#FFE373","white"),
               par('usr')[2], 
               par('usr')[4], xpd=NA);
  dev.off()
}

myPCAplot <- function (plan, info){
  #t(scale(t(plan), center = T, scale = T)) -> plan
  prcomp(t(plan), scale = T, center = T) -> tmp 
  #summary(prcomp(plan)) -> tmp
  fviz_pca_ind(tmp,
             col.ind = "cos2", # Color by the quality of representation
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             ) -> tmp2
  ggplot (as.data.frame(tmp2$data), 
          aes(x=x, y=y, col=info$idade, 
              label = rownames(info),  
              shape=info$sex)) +
    geom_point() +
    geom_text(hjust=0.5,vjust=-0.25) +
    labs(x = paste( "Leading PC1", round(summary(tmp)$importance[2,1]*100,2), "%"), 
         y = paste( "Leading PC2", round(summary(tmp)$importance[2,2]*100,2), "%")) +
   scale_color_continuous("Age") +
   scale_shape_discrete("Sex") -> d
  print (d)
}
```
